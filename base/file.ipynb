小作业
利用百度POI小插件爬取一些poi数据，然后存成txt，再用python读取，编写成一个json形式（列表字典）：
[{'name':'...', 'lng':..., 'lat':..., 'address':'...'},{...},...,]

```python
path = r'D:\user_profile\jiuba.txt'
f = open(path,'r',encoding='gbk')
m = []
s = []
for line in f.readlines():
    st = line.split(':')
    name = st[0]
    st2 = st[1].split(',')
    lng = float(st2[0])
    lat = float(st2[1])
    address = st2[2].strip()
    data = [['name',name],['lng',lng],['lat',lat],['address',address]]
    m.append(dict(data))

print(m)
print('finish')
```

```python
# 文件写入 - write

# 执行后直接创建“test_write.txt”文件，并写入“hello world！”
path = r'D:\user_profile\new.txt'
f = open(path,'w',encoding='utf8')
f.write('Hello Python')
# 执行了f.close()方法后才真正写到本地
f.close()
```

```python
# f.writelines(list)：依次把列表元素写入
path = r'D:\user_profile\new1.txt'
f = open(path,'w',encoding='utf8')
lst = ['a','b','c','d','e','f']
f.writelines(lst)
f.close()
```

```python
# f.writelines(list)：依次把列表元素写入,增加换行
path = r'D:\user_profile\new2.txt'
f = open(path,'w',encoding='utf8')
lst = ['a','b','c','d','e','f']

for i in range(len(lst)):
    lst[i] = lst[i] + '\n'
    
f.writelines(lst)
f.close()
```

'''
【课程8.4】  pickle模块的运用

pickle提供了一个简单的持久化功能。可以将对象以文件的形式存放在磁盘上。

python的pickle模块实现了基本的数据序列和反序列化

通过pickle模块的序列化操作我们能够将程序中运行的对象信息保存到文件中去，永久存储

通过pickle模块的反序列化操作，我们能够从文件中创建上一次程序保存的对象。

pickle.dump() / pickle.load()

'''

```python
# 存储：pickle.dump(obj, file, [,protocol])

import pickle

data = {'a':[1,2,3,4], 'b':('string','abc'), 'c':'hello'}
print(data)

# 以二进制存储
pic = open(r'D:\user_profile\data.pkl','wb')
pickle.dump(data,pic)

# 将字典转换成pkl文件
pic.close()
```

```python
# 读取：pickle.load(file)

f = open(r'D:\user_profile\data.pkl','rb')

# 直接读取pkl文件的数据,该数据是一个字典
st = pickle.load(f)
print(st,type(st))
```
