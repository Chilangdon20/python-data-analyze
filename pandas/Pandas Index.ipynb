```python
'''
【课程2.6】  Pandas数据结构Dataframe：索引

Dataframe既有行索引也有列索引，可以被看做由Series组成的字典（共用一个索引）

选择列 / 选择行 / 切片 / 布尔判断

'''
```

```python
import pandas as pd
import numpy as np
```

```python
# 选择行与列

df = pd.DataFrame(np.random.rand(12).reshape(3,4)*100,
                  index=['one','two','three'],
                  columns=['a','b','c','d'])
print(df)
data1 = df['a']                                         # 按照列名选择列
data2 = df[['a','c']]
print(data1,type(data1))                                # 选择一列输出Series
print(data2,type(data2))                                # 选择多列输出DataFrame
print('-----')

data3 = df.loc['one']                                   # 按照列名选择行 
data4 = df.loc[['one','two']]
print(data3,type(data3))                                # 选择一行输出Series
print(data4,type(data4))                                # 选择多行输出DataFrame
```

```python
# df[] - 选择列
# 一般用于选择列，也可以选择行

df = pd.DataFrame(np.random.rand(12).reshape(3,4)*100,
                  index=['one','two','three'],
                  columns=['a','b','c','d'])
print(df)

data1 = df['a']                                         
data2 = df[['b','c']]

print(data1)
print(data2)

'''
 df[]默认选择列，[]中写列名（所以一般数据colunms都会单独制定，不会用默认数字列名，以免和index冲突）
 单选列为Series，print结果为Series格式
 多选列为Dataframe，print结果为Dataframe格式
 
'''
data3 = df[:1]
print(data3,type(data3))

'''
 df[]中为数字时，默认选择行，且只能进行切片的选择，不能单独选择（df[0]）
 输出结果为Dataframe，即便只选择一行
 df[]不能通过索引标签名来选择行(df['one'])

 注意：df[col]一般用于选择列，[]中写列名
'''
```

```python
# df.loc[] - 按index选择行

df1 = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index=['one','two','three','four'],
                  columns=['a','b','c','d'])
df2 = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                   columns=['a','b','c','d'])
print(df1)
print(df2)
print('-----')

data1 = df1.loc['one']                                    # 单个标签索引，返回Series
data2 = df2.loc[1]
print(data1)
print(data2)
print(f'单标签索引\n')

data3 = df1.loc[['two','three','five']]                   # 多个标签索引，如果标签不存在，则返回NaN
data4 = df2.loc[[3,2,1]]                                  # 索引顺序可变
print(data3)
print(data4)
print(f'多标签索引\n')

data5 = df1.loc['one':'three']                            # 字符串可以做切片对象
data6 = df2.loc[1:3]                                      # 末端包含
print(data5)
print(data6)
print(f'切片索引\n')

''' 注意：df.loc[label]主要针对index选择行，同时支持指定index，及默认数字index'''
```

```python
# df.iloc[] - 按照整数位置（从轴的0到length-1）选择行
# 类似list的索引，其顺序就是dataframe的整数位置，从0开始计

df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index=['one','two','three','four'],
                  columns=['a','b','c','d'])
print(df)
print('-----')

print(df.iloc[0])                                   # 单位置索引
print(df.iloc[-1])
# print(df.iloc[8])                                 # loc索引不同，不能索引超出数据行数的整数位置

print(df.iloc[[0,2]])                               # 多位置索引
print(df.iloc[[3,2,1]])                             # 位置可变

print(df.iloc[1:3])                                 # 切片索引,不包括边界
print(df.iloc[::2])
```

```python
# 布尔型索引

df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index=['one','two','three','four'],
                  columns=['a','b','c','d'])
print(df)
print('1-----')

b1 = df < 20                                    # 索引会对数据每个值进行判断
print(b1,type(b1))
print(df[b1])                                   # 索引结果保留 所有数据：True返回原数据，False返回值为NaN
print('2-----')

b2 = df['a'] > 50                               # 单列做判断
print(b2,type(b2)) 
print(df[b2])                                   # 索引结果保留 单列判断为True的行数据，包括其他列
print('3-----')

b3 = df[['a','b']] > 50                         # 多列做判断
print(b3,type(b3)) 
print(df[b3])                                   # 索引结果保留 所有数据：True返回原数据，False返回值为NaN
print('4-----')

b4 = df.loc[['one','three']] < 50               # 多行做判断
print(b4,type(b4))                               
print(df[b4])                                   # 索引结果保留 所有数据：True返回原数据，False返回值为NaN
```

```python
# 多重索引：比如同时索引行和列
# 先选择列再选择行 —— 相当于对于一个数据，先筛选字段，再选择数据量

df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                  index=['one','two','three','four'],
                  columns=['a','b','c','d'])
print(df)
print('------')

print(df['a'].loc[['one','three']])                    # 选择a列的one，three行
print(df[['b','c','d']].iloc[::2])                     # 选择b,c,d三列,以步长为2选择行 
print(df[df['a']<50].iloc[:2])                         # 选择满足a列小于50的值的前2行
```

```python
	作业1：如图创建Dataframe(4*4，值为0-100的随机数)，通过索引得到以下值
① 索引得到b，c列的所有值
② 索引得到第三第四行的数据
③ 按顺序索引得到two，one行的值
④ 索引得到大于50的值
```

```python
df = pd.DataFrame(np.random.rand(16).reshape(4,4)*100,
                 columns=['a','b','c','d'],
                 index=['one','two','three','four'])
print(df)

print(f"b,c列的值是:\n{df[['b','c']]}")
print(f"第三行,第四行的值是:\n{df.loc[['three','four']]}")
print(f"顺序索引的值是:\n{df.iloc[[1,0]]}")
print(f"大于50值是:\n{df[df>50]}")
```
