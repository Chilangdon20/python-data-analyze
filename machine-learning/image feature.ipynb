## 图像 shape

```python
cat = io.imread('D:/user_profile/python/machine-learning/dataset/cat.png')  # Reading, saving, and displaying images and video.
dog = io.imread('D:/user_profile/python/machine-learning/dataset/dog.png')

df = pd.DataFrame(['Cat','Dog'],columns=['Image'])

print(cat.shape,dog.shape)
```

```python
cat                                                                          # 三维数组,0-255,越小的值代表越暗，越大的值越亮
 
```

```python
fig = plt.figure(figsize=(10,4))            # 展示图片
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(cat)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(dog)
```

```python
# 单色图

dog_r = dog.copy()
dog_r[:,:,1] = dog_r[:,:,2] = 0           # 红色  set G,B pixels = 0
dog_g = dog.copy()
dog_g[:,:,0] = dog_g[:,:,2] = 0           # 绿色  set R,B pixels = 0
dog_b = dog.copy()
dog_b[:,:,0] = dog_b[:,:,1] = 0           # 蓝色  set R,G pixels = 0

plot_image = np.concatenate((dog_r,dog_g,dog_b),axis=1)
plt.figure(figsize=(12,4))
plt.imshow(plot_image)

dog_r
```

```python
# 灰度图

fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat,cmap='gray')
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(dog,cmap='gray')
```
