```python
import numpy as np
```

```python
# 随机数生成

samples = np.random.sample(size=(4,4))               # 生成一个标准正太分布的4*4样本值
print(samples)                 
```

```python
import matplotlib.pyplot as plt                # 导入matplotlib模块，用于图表辅助分析
% matplotlib inline

# numpy.random.rand(d0, d1, ..., dn)：生成一个[0,1)之间的随机浮点数或N维浮点数组 —— 均匀分布
a = np.random.rand()
print(a,type(a))                                # 生成一个随机浮点数

b = np.random.rand(4)
print(b,type(b))                                # 生成形状为4的一维数组

c = np.random.rand(2,3)                         # 生成形状为2*3的二维数组
print(c,type(c))

samp1 = np.random.rand(1000)                    # 生成1000个均匀分布的样本值
samp2 = np.random.rand(1000)
plt.scatter(samp1,samp2)
```

```python
#  numpy.random.randn(d0, d1, ..., dn)：生成一个浮点数或N维浮点数组 —— 正态分布

samp1 = np.random.randn(1000)
samp2 = np.random.randn(1000)
plt.scatter(samp1,samp2)
```

```python
# numpy.random.randint(low, high=None, size=None, dtype='l')：生成一个整数或N维整数数组
# 若high不为None时，取[low,high)之间随机整数，否则取值[0,low)之间随机整数，且high必须大于low 
# dtype参数：只能是int类型  

print(np.random.randint(2))               # low=2：生成1个[0,2)之间随机整数  

print(np.random.randint(2,size=5))        # low=2,size=5 ：生成5个[0,2)之间随机整数

print(np.random.randint(2,6,size=5))      # low=2,high=6,size=5：生成5个[2,6)之间随机整数 

print(np.random.randint(2,size=(2,3)))    # low=2,size=(2,3)：生成一个2x3整数数组,取数范围：[0,2)随机整数

print(np.random.randint(2,6,size=(2,3)))  # low=2,high=6,size=(2,3)：生成一个2*3整数数组,取值范围：[2,6)随机整数  
```
